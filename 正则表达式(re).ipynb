```python
import re
```

```python
re1=re.compile(r'[_a-zA-Z]\w*')#声明一个标识符规则的正则表达式
```

```python
#从一个字符串的开始位置起匹配正则表达式，返回match对象
m1=re1.match('acef22_')
m1.group(0)
m1[0]
```

```python
type(m1)
```

```python
m2=re1.match('2eewr')    #当未匹配到时，返回None;匹配到时返回 class mathch(特定的类对象)
if(m2==None):
    print('No find!')
```

```python
m2.string
```

```python
help(re1.match)
```

```python
#在一个字符串中搜索匹配正则表达式的第一个位置返回match对象
m3=re1.search('2eewr')
m3[0]
```

```python
#搜索字符串，以列表类型返回全部能匹配的子串
m4=re1.findall('2eewr..ew2erq')
m4
```

```python
help(re1.findall)
```

```python
m8=re.search(r'[A-Z]*','asZds')
m8.group(0)
```

```python
m9=re.search(r'[A-Z]*','asZds',flags=re.I)
print(m9)
```

```python
help(m8.group)
```

```python
#*默认是采用贪婪匹配算法的，但加上？后就是基于最小匹配
m9=re.search(r'[A-Z]*?','asZds',flags=re.I)
print(m9)
```

```python
matchLongest=re.search(r'PY.*N','PYANBNCNDN')
matchLongest.group(0)
```

```python
matchShortest=re.search(r'PY.*?N','PYANBNCNDN')
matchShortest.group(0)
```
