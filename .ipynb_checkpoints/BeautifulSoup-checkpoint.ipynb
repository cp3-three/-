## 多种遍历方式


```python
p.contents
```

```python
len(p.contents)
```

```python
for i in p.contents:
    print(type(i))
    print(i)
    print()
```

```python
for i in p.children:
    print(type(i))
    print(i)
    print()
```

```python
for i in p.descendants:
    print(type(i))
    print(i)
    print()
```

```python
#树的中序遍历
for i in soup.descendants:
    print(type(i))
    print(i)
    print()
```

```python
soup.a
```

```python
#遍历所有先辈节点，包括soup本身，所以要区别判断
for parent in soup.a.parents:
    if parent is None:
        print(parent)
    else:
        print(parent.name)
```

```python
soup.a.parent
```

```python
for i in soup.a.parent:
    print(type(i))
    print(i)
    print()
```

```python
#平行遍历发生在同一个父节点下的各节点间

#遍历后续节点
for sibling in soup.a.next_siblings:
    print(sibling)
```

```python
#遍历前续节点
for sibling in soup.a.previous_siblings:
    print(sibling)
```
