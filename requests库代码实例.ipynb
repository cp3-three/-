```python
#获取html网页内容时，由于网页连接的不确定性，必须对请求连接的异常进行处理
def getHTMLText(url):
    try:
        r=requests.get(url)
        r.raise_for_status() #若返回码不是200，会抛出异常
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return "产生异常"
```

```python
def getHTMLText(url):
    try:
        #由于网站对爬虫的限制，故应设个代理服务器
        kv={'user-agent':'Mozilla/78.0.2'}
        r=requests.get(url,kv)
        r.raise_for_status() #若返回码不是200，会抛出异常
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return "产生异常"
```

```python
#搜索引擎关键字提交接口,并爬取搜索结果
keyword="python"
try:
    kv={"wd":keyword}     #百度的用wd，360的用q
    r=requests.get("http://www.baidu.com/s",params=kv)
    print(r.request.url)
    r.raise_for_status()
    r.encoding=r.apparent_encoding
    print(len(r.text))
except:
    print("爬取失败")
```
